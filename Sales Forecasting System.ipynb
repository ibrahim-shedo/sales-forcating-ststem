## EDA: Sales Over Time, Monthly Trends, Seasonality

```python
# Convert date column correctly (day-first format)
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
```

```python
## Convert date column correctly (day-first format) for Ship Date column too
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=True)
```

```python
## cheking order date data type 
df["Order Date"].head()
df.info()
```

```python
import matplotlib.pyplot as plt

# Group sales by date
sales_over_time = df.groupby("Order Date")["Sales"].sum()

# Plot trend
plt.figure(figsize=(12,5))
plt.plot(sales_over_time)

plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert to monthly datetime
df["Month"] = df["Order Date"].dt.to_period("M").dt.to_timestamp()

# Group by month
monthly_sales = df.groupby("Month")["Sales"].sum()

# Plot
plt.figure(figsize=(12,5))
plt.plot(monthly_sales)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# Create month number
df["Month_num"] = df["Order Date"].dt.month

# Average sales per month
seasonality = df.groupby("Month_num")["Sales"].mean()

# Plot
plt.figure(figsize=(10,5))
plt.bar(seasonality.index, seasonality.values)

plt.title("Seasonality Pattern (Average Sales per Month)")
plt.xlabel("Month")
plt.ylabel("Average Sales")
plt.show()
```

## BASELINE FORECASTING MODEL

```python
# Ensure date is datetime
df["Order Date"] = pd.to_datetime(df["Order Date"])

# Sort by date
df = df.sort_values("Order Date")

# Aggregate sales by date
daily_sales = df.groupby("Order Date")["Sales"].sum()
```

```python
# Create rolling mean (7-day moving average)
moving_avg = daily_sales.rolling(window=7).mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

# Actual sales
plt.plot(daily_sales, label="Actual Sales")

# Forecast (smoothed trend)
plt.plot(moving_avg, label="7-Day Moving Average", color="red")

plt.title("Sales Forecast Baseline (Moving Average)")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

The moving average shows the underlying sales trend by smoothing short-term fluctuations. This helps the business understand whether sales are generally increasing or declining over time.

```python
#Prepare data ds = date
# y = value (sales)

from prophet import Prophet
import pandas as pd

# Prepare dataset
prophet_df = df.groupby("Order Date")["Sales"].sum().reset_index()

# Rename columns for Prophet
prophet_df.columns = ["ds", "y"]
```

```python
from prophet import Prophet
print("Prophet is working")
```

```python
#Train model
model = Prophet()

# Fit model
model.fit(prophet_df)
```

```python
#Create future dates
#Example: predict next 30 days
future = model.make_future_dataframe(periods=30)

forecast = model.predict(future)
```

```python
#Visualize forecast
fig = model.plot(forecast)
fig.show()
```

```python
#Show trend + seasonality
fig2 = model.plot_components(forecast)
fig2.show()
```
